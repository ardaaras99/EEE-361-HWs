```python
def svd_init(U,S,V_T,r = 49):
    W = np.zeros((361,r))
    H = np.zeros((r,2429))
    V = V_T.T

    for i in range(r):
    
        uk = U[:,i].reshape(-1,1)
        vk_t = V[:,i].reshape(1,-1)

        uk_pos = np.maximum(uk,0)
        uk_neg = np.maximum(-1*uk,0)

        vk_t_pos = np.maximum(vk_t,0)
        vk_t_neg = np.maximum(-1*vk_t,0)
        
        Mk_plus = uk_pos @ vk_t_pos#(361,2429)
        Mk_minus = uk_neg @ vk_t_neg#(361,2429)
        
        if np.linalg.norm(Mk_plus,ord='fro') >= np.linalg.norm(Mk_minus,ord = 'fro'):
            W[:,i] = (uk_pos / np.linalg.norm(uk_pos)).reshape(361)
            H[i,:] = (S[i]*np.linalg.norm(uk_pos) * vk_t_pos).reshape(2429)
        else:
            W[:,i] = (uk_neg / np.linalg.norm(uk_neg)).reshape(361)
            H[i,:] = (S[i]*np.linalg.norm(uk_neg) * vk_t_neg).reshape(2429)
       
    return W,H
```

```python
def hals_update(X,H,W):
    W_new = np.copy(W)
    for l in range(W_new.shape[1]):
        norm_mat = H @ H[l,:]
        a = X @ H[l,:].T - W_new @ norm_mat + W_new[:,l] * (norm_mat[l])
        b = np.linalg.norm(H[l,:]) ** 2
        c = a/b
        c = np.maximum(0,c).reshape(W_new.shape[0])
        W_new[:,l] = c
    return W_new
```

```python
def coordinate_descent(X,W_0,H_0, delta = 0.01):
    
    W_prev = hals_update(X, H_0, W_0)
    H_prev = hals_update(X.T,W_prev.T,H_0.T).T

    stop = delta*np.linalg.norm(W_prev - W_0) 
    stop2 = delta*np.linalg.norm(H_prev - H_0)
    
    norm = np.linalg.norm(X)
    W_new = np.zeros_like(W_prev)
    H_new = np.zeros_like(H_prev)
    err_acc = []

    curr_diff1 = np.linalg.norm(W_new - W_prev)
    curr_diff2 = np.linalg.norm(H_new - H_prev)
    i = 0
    print("Stops are " , stop,stop2)
    while (curr_diff1 > stop) and (curr_diff2 > stop2):
        
        W_new = hals_update(X,H_prev,W_prev)
        H_new = hals_update(X.T,W_new.T,H_prev.T).T
        
        curr_diff1 = np.linalg.norm(W_new - W_prev)
        curr_diff2 = np.linalg.norm(H_new - H_prev)
        
        err = np.linalg.norm( X - W_new @ H_new ) / norm
        err_acc.append(err)
        
        W_prev, H_prev = W_new, H_new
        
    return W_new,H_new,np.array(err_acc)
```

```python
W_0,H_0 = svd_init(u,s,vt)
W,H,err_accum = coordinate_descent(X_train, W_0, H_0)
```

```python
def add_noise(X,n):
    a,b,c = X.shape
    X_noisy = np.zeros((a,b,c))
    for i in range(X.shape[0]):
        X_noisy[i,:,:] = np.minimum(X[i,:,:]+n*np.random.rand(19,19),255)
    return X_noisy
```

```python
def vec(X):
    mylist = []
    for image in X:
        image_f = image.reshape(361,order = 'F')
        mylist.append(image_f)
    return np.array(mylist).T
```

```python
def reconstruct_from_NMF(W,vectorized):
    mylist = []
    for i in range(vectorized.shape[1]):
        #For each image
        yk_hat = np.zeros((361))
        beta = np.linalg.lstsq(W,vectorized[:,i],rcond = None)[0].reshape(-1,1)
        yk_hat = np.matmul(W,beta).reshape(361)
        mylist.append(yk_hat)
    return np.array(mylist).T
```

```python
def mat(y_hat,size):
    mylist = []
    for i in range(y_hat.shape[1]):
        mylist.append(y_hat[:,i].reshape(size,size,order = 'F'))
    return np.array(mylist)
```

```python
def average_error(images_recon,X):
    #They should be same size 472,19,19
    err = np.zeros((472))
    for i in range(images_recon.shape[0]):
        err[i] = np.linalg.norm(X[i,:,:] - images_recon[i,:,:],ord = 'fro')
    return err, np.average(err)
```

```python
def reconstruct_from_SVD(U_train,vec,r_svd):
    mylist = []
    a = U_train[:,:r_svd]
    yk_svd = a @ a.T @ vec
    return yk_svd
```

```python
#Now we need to reconstruct to noisy version of the images for 1 <= r <= 100.
#Meaning that we want to observe the affect of adding new columns from the U matrix to the error
err_n1 = np.zeros((100))
err_n2 = np.zeros((100))
err_n3 = np.zeros((100))

for i in range(0,100):
    vec_n1 = vec(X_test_n1)
    rec_n1 = reconstruct_from_SVD(U_train,vec_n1,i+1)
    images_recon_n1_svd = mat(rec_n1,19)
    _,err_n1[i] = average_error(images_recon_n1_svd, X_test)
    
    vec_n2 = vec(X_test_n2)
    rec_n2 = reconstruct_from_SVD(U_train,vec_n2,i+1)
    images_recon_n2_svd = mat(rec_n2,19)
    _,err_n2[i] = average_error(images_recon_n2_svd, X_test)
    
    vec_n3 = vec(X_test_n3)
    rec_n3 = reconstruct_from_SVD(U_train,vec_n3,i+1)
    images_recon_n3_svd = mat(rec_n3,19)
    _,err_n3[i] = average_error(images_recon_n3_svd, X_test)
```
