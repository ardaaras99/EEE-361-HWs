```python
def initialize_W_H(U,S,V_T,r):
    W = np.zeros((361,r))
    H = np.zeros((r,2429))
    V = V_T.transpose()

    for i in range(r):
    
        u_k = U[:,i].reshape(-1,1)# (361,1)
        v_k_t = V[:,i].reshape(1,-1)#(1,2429)

        u_k_plus = np.maximum(u_k,0)#(361,1)
        u_k_minus = np.maximum(-1*u_k,0)#(361,1)

        v_k_t_plus = np.maximum(v_k_t,0)#(1,2429)
        v_k_t_minus = np.maximum(-1*v_k_t,0)#(1,2429)
        
        M_k_plus = np.matmul(u_k_plus,v_k_t_plus)#(361,2429)
        M_k_minus = np.matmul(u_k_minus,v_k_t_minus)#(361,2429)
        
        if np.linalg.norm(M_k_plus,ord = 'fro') >= np.linalg.norm(M_k_minus,ord = 'fro'):
            W[:,i] = (u_k_plus / np.linalg.norm(u_k_plus)).reshape(361)
            H[i,:] = (S[i]*np.linalg.norm(u_k_plus) * v_k_t_plus).reshape(2429)
        else:
            W[:,i] = (u_k_minus / np.linalg.norm(u_k_minus)).reshape(361)
            H[i,:] = (S[i]*np.linalg.norm(u_k_minus) * v_k_t_minus).reshape(2429)
       
    return W,H
```

```python
def update(X,H,W):
    for l in range(W.shape[1]):
        summation = np.zeros(W.shape[0]).reshape(-1,1)
        for  k in range(W.shape[1]):
            if(k != l):
                summation += W[:,k].reshape(-1,1) * np.matmul(H[k,:].reshape(1,-1),H[l,:].reshape(-1,1))
        numerator = np.matmul(X,H[l,:].reshape(-1,1)) - summation
        W[:,l] = np.nan_to_num(np.maximum(0,numerator/((np.linalg.norm(H[l,:]))**2)).reshape(W.shape[0]))   
    return W
```

```python
def coordinate_descent(X,number_of_iteration, r_nmf ):
    U,S,V_T = np.linalg.svd(X_train,full_matrices = False)
    W_initial, H_initial = initialize_W_H(U,S,V_T,r_nmf )
    W = W_initial
    H_T = H_initial.transpose()
    error_convergence = np.zeros((number_of_iteration))
    for i in range(number_of_iteration):
        W = update(X,H_T.transpose(),W)
        H_T = update(X.transpose(),W.transpose(),H_T)
        error_convergence[i] = np.linalg.norm(X_train-np.matmul(W,H_T.T))
    return W,H_T,error_convergence
```

```python
W_new,H_T_new,err_convergence = coordinate_descent( X_train,number_of_iteration= 200,r_nmf = 30)
```

```python
def vec(X):
    X_copy = X
    mylist = []
    for i in range(X.shape[0]):
        mylist.append(X[i,:,:].flatten('F'))
    return np.array(mylist).transpose()
```

```python
def reconstruct_SVD(U_train,vectorized,r_SVD):
    mylist = []
    for i in range(vectorized.shape[1]):
        #For each image
        yk_hat = np.zeros((361))
        for j in range(r_SVD):
            yk_hat += np.matmul(vectorized[:,i].reshape(1,-1),U_train[:,j].reshape(-1,1)).reshape(1)*U_train[:,j].reshape(361)
        mylist.append(yk_hat)
    return np.array(mylist).transpose()
```

```python
def reconstruct_NMF(W,vectorized):
    mylist = []
    for i in range(vectorized.shape[1]):
        #For each image
        yk_hat = np.zeros((361))
        beta = np.linalg.lstsq(W,vectorized[:,i],rcond = None)[0].reshape(-1,1)
        yk_hat = np.matmul(W,beta).reshape(361)
        mylist.append(yk_hat)
    return np.array(mylist).transpose()
```

```python
def mat(y_hat,size):
    mylist = []
    for i in range(y_hat.shape[1]):
        mylist.append(y_hat[:,i].reshape(size,size,order = 'F'))
    return np.array(mylist)
```

```python
def average_error(images_recon,X):
    #They should be same size 472,19,19
    Ek = np.zeros((472))
    for i in range(images_recon.shape[0]):
        Ek[i] = np.linalg.norm(X[i,:,:] - images_recon[i,:,:],ord = 'fro')
    return Ek, np.average(Ek)
```
