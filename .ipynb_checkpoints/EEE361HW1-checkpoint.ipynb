```python
from skimage.io import imread
import os
import cv2 as cv
import numpy as np
import math
import matplotlib.pyplot as plt

from sklearn.decomposition import NMF

def load_images_from_folder_flattened(folder):
    images = []
    for filename in os.listdir(folder):
        img = imread(os.path.join(folder,filename))
        img = img.flatten('F')
        if img is not None:
            images.append(img)
    return images
cwd = os.getcwd()
folder= cwd + "/Dataset/train"
images = load_images_from_folder_flattened(folder)
X_train = np.array(images).transpose()
```

```python
def initialize_W_H(U,S,V_T,r = 50):
    W = np.zeros((361,r))
    H = np.zeros((r,2429))
    V = V_T.transpose()

    for i in range(r):
    
        u_k = U[:,i].reshape(-1,1)# (361,1)
        v_k_t = V[:,i].reshape(1,-1)#(1,2429)

        u_k_plus = np.maximum(u_k,0)#(361,1)
        u_k_minus = np.maximum(-1*u_k,0)#(361,1)

        v_k_t_plus = np.maximum(v_k_t,0)#(1,2429)
        v_k_t_minus = np.maximum(-1*v_k_t,0)#(1,2429)
        
        M_k_plus = np.matmul(u_k_plus,v_k_t_plus)#(361,2429)
        M_k_minus = np.matmul(u_k_minus,v_k_t_minus)#(361,2429)
        
        if np.linalg.norm(M_k_plus,ord = 'fro') >= np.linalg.norm(M_k_minus,ord = 'fro'):
            W[:,i] = (u_k_plus / np.linalg.norm(u_k_plus)).reshape(361)
            H[i,:] = (S[i]*np.linalg.norm(u_k_plus) * v_k_t_plus).reshape(2429)
        else:
            W[:,i] = (u_k_minus / np.linalg.norm(u_k_minus)).reshape(361)
            H[i,:] = (S[i]*np.linalg.norm(u_k_minus) * v_k_t_minus).reshape(2429)
       
    return W,H
```

```python
W_init,H_init = initialize_W_H(U,S,VT)

```
