```python
def vec(X):
    mylist = []
    for image in X:
        image_f = image.reshape(361,order = 'F')
        mylist.append(image_f)
    return np.array(mylist).T
```

```python
def reconstruct_from_SVD(U_train,vec,r_svd):
    mylist = []
    for k in range(vec.shape[1]):
        yk_svd = np.zeros(361).reshape(361,1)
        yk_hat = vec[:,k].reshape(361,1)
        yk_hat = yk_hat.T
        for p in range(r_svd):
            u_p = U_train[:,p].reshape(361,1)
            yk_svd = yk_svd + (yk_hat @ u_p) * u_p
        yk_svd = yk_svd.reshape(361)
        mylist.append(yk_svd)
    return np.array(mylist).T
```

```python
def reconstruct_NMF(W,vectorized):
    mylist = []
    for i in range(vectorized.shape[1]):
        #For each image
        yk_hat = np.zeros((361))
        beta = np.linalg.lstsq(W,vectorized[:,i],rcond = None)[0].reshape(-1,1)
        yk_hat = np.matmul(W,beta).reshape(361)
        mylist.append(yk_hat)
    return np.array(mylist).transpose()
```

```python
def mat(y_hat,size):
    mylist = []
    for i in range(y_hat.shape[1]):
        mylist.append(y_hat[:,i].reshape(size,size,order = 'F'))
    return np.array(mylist)
```

```python
def average_error(images_recon,X):
    #They should be same size 472,19,19
    Ek = np.zeros((472))
    for i in range(images_recon.shape[0]):
        Ek[i] = np.linalg.norm(X[i,:,:] - images_recon[i,:,:],ord = 'fro')
    return Ek, np.average(Ek)
```

```python
vec_X_n1 = vec(X_test_n1)
print(X_test_n1.shape)
print(vec_X_n1.shape)
```

```python
y_hat = reconstruct_from_SVD(U_train, vec_X_n1 ,30) #every colmn of y_hat corresponds to single image
print(y_hat.shape)
images_recon = mat(y_hat,19)  #reconstructing images to original form
print(images_recon.shape,X_test.shape)
_,av_error = average_error(images_recon, X_test)
```
