```python
from skimage.io import imread
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import NMF

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images

def load_images_from_folder2(folder):
    images = []
    for filename in os.listdir(folder):
        img = imread(os.path.join(folder,filename))
        img = img.flatten('F')
        if img is not None:
            images.append(img)
    return images

cwd = os.getcwd()
folder= cwd + "/Dataset/train"
images = load_images_from_folder2(folder)
X_train = np.array(images).transpose()

cwd = os.getcwd()
folder= cwd + "/Dataset/test"
images = load_images_from_folder(folder)
X = np.array(images)
```

```python
def vec(X):
    X_copy = X
    mylist = []
    for i in range(X.shape[0]):
        mylist.append(X[i,:,:].flatten('F'))
    return np.array(mylist).transpose()
```

```python
def reconstruct_SVD(U_train,vectorized,r_SVD):
    mylist = []
    for i in range(vectorized.shape[1]):
        #For each image
        yk_hat = np.zeros((361))
        for j in range(r_SVD):
            yk_hat += np.matmul(vectorized[:,i].reshape(1,-1),U_train[:,j].reshape(-1,1)).reshape(1)*U_train[:,j].reshape(361)
        mylist.append(yk_hat)
    return np.array(mylist).transpose()
```

```python
def reconstruct_NMF(W,vectorized):
    mylist = []
    for i in range(vectorized.shape[1]):
        #For each image
        yk_hat = np.zeros((361))
        beta = np.linalg.lstsq(W,vectorized[:,i],rcond = None)[0].reshape(-1,1)
        yk_hat = np.matmul(W,beta).reshape(361)
        mylist.append(yk_hat)
    return np.array(mylist).transpose()
```

```python
def mat(y_hat,size):
    mylist = []
    for i in range(y_hat.shape[1]):
        mylist.append(y_hat[:,i].reshape(size,size,order = 'F'))
    return np.array(mylist)
```

```python
def average_error(images_recon,X):
    #They should be same size 472,19,19
    Ek = np.zeros((472))
    for i in range(images_recon.shape[0]):
        Ek[i] = np.linalg.norm(X[i,:,:] - images_recon[i,:,:],ord = 'fro')
    return Ek, np.average(Ek)
```

```python
vec_ten = vec(X_noised_tfive)
y_hat = reconstruct_SVD(U_train, vec_ten,30)
images_recon = mat(y_hat,19)
_,av_error = average_error(images_recon, X)
```
